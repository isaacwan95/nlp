```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('nlp.csv')
data.columns
```

```python
a = sum(data['price']/len(data['price']))
l = []
for i in data['price']:
    if i > a:
        i = "Expensive"
    else:
        i = "Cheap"
    l.append(i)
data['pricelevel'] = l
data.columns
```

```python
data.brand.unique()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)
```

```python
# Make a separate list for each airline
x1 = list(data[data['brand'] == 'Nokia']['price'])
x2 = list(data[data['brand'] == 'Samsung']['price'])
x3 = list(data[data['brand'] == 'HTC']['price'])
x4 = list(data[data['brand'] == 'Sony']['price'])
x5 = list(data[data['brand'] == 'BlackBerry']['price'])
x6 = list(data[data['brand'] == 'LG']['price'])
x7 = list(data[data['brand'] == 'Apple']['price'])

colors = ['#E69F00', '#56B4E9', '#F0E442', '#009E73', '#D55E00', 'y', 'm']
names = ['Nokia', 'Samsung', 'HTC', 'Sony', 'BlackBerry', 'LG', 'Apple']

plt.hist([x1, x2, x3, x4, x5, x6, x7], stacked=True, color = colors, label=names)

# Plot formatting
plt.legend()
plt.xlabel('Price')
plt.ylabel('Frequency of price')
plt.title('Price distribution for each brand')
```

price distribution

```python
# Density Plot and Histogram of all arrival delays
sns.distplot(data['price'],  kde=True, 
            color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```

price distribution for each brand

```python

# Assign colors for each airline and the names
colors = ['#E69F00', '#56B4E9', '#F0E442', '#009E73', '#D55E00', 'y', 'm']
names = ['Nokia', 'Samsung', 'HTC', 'Sony', 'BlackBerry', 'LG', 'Apple']
         
# Make the histogram using a list of lists
# Normalize the flights and assign colors and names
plt.hist([x1, x2, x3, x4, x5, x6, x7], color = colors, label=names)

# Plot formatting
plt.legend()
plt.xlabel('Price')
plt.ylabel('Frequency of price')
plt.title('Price distribution for each brand')
```

```python
# List of 7 brands to plot
names = ['Nokia', 'Samsung', 'HTC', 'Sony', 'BlackBerry', 'LG', 'Apple']

# Iterate through the 7 brand
for name in names:
    # Subset to the brand
    subset = data[data['brand'] == name]
    
    # Draw the density plot
    sns.distplot(subset['price'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = name)

    
# Plot formatting
plt.legend(prop={'size': 16}, title = 'brand')
plt.xlabel('Price')
plt.ylabel('Frequency of price')
plt.title('Price distribution for each brand')
```

rating for diiferent price level

```python
levels = ['Expensive', 'Cheap']

for level in levels:
    subset = data[data['pricelevel'] == level]
    sns.distplot(subset['overall'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = level)

    
# Plot formatting
plt.legend(prop={'size': 16}, title = 'brand')
plt.xlabel('rating')
plt.ylabel('Frequency of rating')
plt.title('Rating distribution for different price')
```
